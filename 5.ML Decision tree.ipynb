```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier # DecisionTreeClassifier fits a tree 
from sklearn.metrics import accuracy_score
from collections import Counter
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
data =pd.read_csv("iris.csv")
data.head(3)
```

```python
data.shape
```

```python
Counter(data.target)
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,:-1]
y = data.target
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size = 0.3,random_state=10)#test_size
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
pd.crosstab(y_test,y_predict)
```

```python
rcParams['figure.figsize']=18,12
dtc = tree.DecisionTreeClassifier(random_state=10)

dtc = dtc.fit(X,y)
tree.plot_tree(dtc,
                  class_names=['setosa','versicolor','virginica'],
                  filled=True,#filled our graph
                  rounded=True,#for visualize the box
                  max_depth=2,#pruning
                  feature_names=['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```
