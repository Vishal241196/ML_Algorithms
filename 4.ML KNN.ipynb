1. preprocessing:-
            -> It is technique that is used to convert the raw data set into more suitable data set
            
2. LabelEncoder:-
            -> Used to normalize labels.
            -> Used to transform non-numerical label to numerical label.
            -> It encode target labels with value B/W 0 and n_classes-1.

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier #KNeighborsClassifier fits a neighbors
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from collections import Counter
```

```python
data = pd.read_csv('car_evaluation.csv')
data.head()
```

```python
Counter(data.outcome)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X =data.iloc[:,:-1]
X.head(3)
```

```python
y = data.outcome
```

```python
enc = LabelEncoder()
X.buying = enc.fit_transform(X.buying)
X.maint = enc.fit_transform(X.maint)
X.lug_boot = enc.fit_transform(X.lug_boot)
X.safety = enc.fit_transform(X.safety)
X.head(3)
```

```python
pd.get_dummies(data)# used to convert categorical variable into dummy/indicator variables.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=10)
```

```python
model = KNeighborsClassifier(n_neighbors=5,metric='euclidean')
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
print(Counter(y_test))
pd.crosstab(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```
